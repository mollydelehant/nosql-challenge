5. Some of the number values are stored as strings, when they should be stored as numbers.

Use `update_many` to convert `latitude` and `longitude` to decimal numbers.

```python
# Change the data type from String to Decimal for longitude and latitude
establishments.update_many({}, [ {'$set': {"geocode.longitude" : {'$toDecimal' : '$geocode.longitude'}}}])
establishments.update_many({}, [ {'$set': {"geocode.latitude" : {'$toDecimal' : '$geocode.latitude'}}}])
```

Use `update_many` to convert `RatingValue` to integer numbers.

```python
# Set non 1-5 Rating Values to Null
non_ratings = ["AwaitingInspection", "Awaiting Inspection", "AwaitingPublication", "Pass", "Exempt"]
establishments.update_many({"RatingValue": {"$in": non_ratings}}, [ {'$set':{ "RatingValue" : None}} ])
```

```python
# Change the data type from String to Integer for RatingValue
change_data = [{'$set': {'RatingValue': {'$toDouble': '$RatingValue'}}}]
establishments.update_many({}, change_data)
```

### 2. Which establishments in London have a `RatingValue` greater than or equal to 4?

```python
# Find the establishments with London as the Local Authority and has a RatingValue greater than or equal to 4.
London = '.*London.*'
query = {
    'LocalAuthorityName': {'$regex': London},
    'RatingValue': {'$gte': 4}
}

# Use count_documents to display the number of documents in the result
count_london = establishments.count_documents(query)
print("Number of establishments that have a RatingValue greater than or equal to 4:", count_london)

# Display the first document in the results using pprint
document = establishments.find_one(query)
pprint(document)
```

```python
# Convert the result to a Pandas DataFrame
cursor_london = establishments.find(query)
df_london = pd.DataFrame(list(cursor_london))

# Display the number of rows in the DataFrame
count_london = len(df_london.index)
print("Number of rows:", count_london)

# Display the first 10 rows of the DataFrame
df_london.head(10)
```

### 3. What are the top 5 establishments with a `RatingValue` rating value of 5, sorted by lowest hygiene score, nearest to the new restaurant added, "Penang Flavours"?

```python
degree_search = 0.01
latitude = 51.490142
longitude = 0.08384

query = {
    'geocode.latitude': {'$lte': latitude + degree_search, '$gte': latitude - degree_search},
    'geocode.longitude': {'$lte': longitude + degree_search, '$gte': longitude - degree_search},
    'RatingValue': 5
}

sort = [('scores.Hygiene', 1)]
limit = 5

# Print the results
for result in establishments.find(query).sort(sort).limit(limit):
    pprint(result)

```

```python
# Convert result to Pandas DataFrame
pd.DataFrame(establishments.find(query).sort(sort).limit(limit))
```

### 4. How many establishments in each Local Authority area have a hygiene score of 0?

```python
# Create a pipeline that:
# 1. Matches establishments with a hygiene score of 0
query_0 = {'$match': {'scores.Hygiene': 0}}
# 2. Groups the matches by Local Authority
query_LA = {'$group': {'_id': '$LocalAuthorityName', 'count': {'$sum': 1}}}
# 3. Sorts the matches from highest to lowest
sort = {'$sort': {'count': -1}}

# Apply the pipeline to the collection
pipeline = [query_0, query_LA, sort]
result = list(establishments.aggregate(pipeline))

# Print the number of documents in the result
print('Number of establishments in each local authority area with a hygiene score of 0:', len(result))

# Print the first 10 results
for i, item in enumerate(result[:10], 1):
    print(f"{i}. {item['_id']}: {item['count']}")
```
